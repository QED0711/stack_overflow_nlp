```python
so_data[so_data.tags.str.contains("git")].tags
```

```python
unique_tags = {}
for tag in so_data.tags:
    individual_tags = tag.split("|")
    for t in individual_tags:
        if t in unique_tags:
            unique_tags[t] += 1
        else:
            unique_tags[t] = 1
        
unique_tags
```

```python
tags = sorted(unique_tags.items(), key=lambda kv: kv[1], reverse=True)
```

```python
pd.DataFrame(tags, columns=("tag", "count")).to_csv("stack_overfow_tags.csv", index=False)
```
